```python
cashRewardsFilePath = '..\data\BOfA_CreditCard_CashRewards_ExportData.csv'
travelRewardsFilePath = '..\data\BOfA_CreditCard_TravelRewards_ExportData.csv'
bofaAccountsFilePath = '..\data\BOfA_ExportData.csv'
discoverFilePath = '..\data\Discover-AllAvailable-20190429.csv'
bofaCashRewardsTransactions = pd.read_csv(cashRewardsFilePath)
bofaTravelRewardsTransactions = pd.read_csv(travelRewardsFilePath)
bofaAccountTransactions = pd.read_csv(bofaAccountsFilePath)
discoverTransactions = pd.read_csv(discoverFilePath)
```

```python
print(bofaCashRewardsTransactions.head())
print(bofaTravelRewardsTransactions.head())
print(bofaAccountTransactions.head())
print(discoverTransactions.head())
```

```python
bofaCashRewardsTransactions.head()
```

```python
bofaCashRewardsTransactions.Amount = bofaCashRewardsTransactions.Amount.str.replace(',','').astype(float)
bofaCashRewardsTransactions.Amount = bofaCashRewardsTransactions.Amount * -1
bofaCashRewardsTransactions.rename(columns={'Simple Description': 'Description'},inplace=True)
bofaCashRewardsTransactions.Description.fillna(bofaCashRewardsTransactions['Original Description'],inplace=True)
bofaCashRewardsTransactions.head()
```

```python
bofaTravelRewardsTransactions.Amount = bofaTravelRewardsTransactions.Amount.str.replace(',','').astype(float)
bofaTravelRewardsTransactions.Amount = bofaTravelRewardsTransactions.Amount * -1.0
bofaTravelRewardsTransactions.rename(columns={'Simple Description': 'Description'},inplace=True)
bofaTravelRewardsTransactions.Description.fillna(bofaTravelRewardsTransactions['Original Description'],inplace=True)
bofaTravelRewardsTransactions.head()
```

```python
bofaAccountTransactions.Amount = bofaAccountTransactions.Amount.str.replace(',','').astype(float)
bofaAccountTransactions.Amount = bofaAccountTransactions.Amount * -1.0
bofaAccountTransactions.rename(columns={'Simple Description': 'Description'},inplace=True)
bofaAccountTransactions.Description.fillna(bofaAccountTransactions['Original Description'],inplace=True)
bofaAccountTransactions.head()
```

```python
bofaAccountTransactions[bofaAccountTransactions.Category == 'Other Income']
```

```python
discoverTransactions.rename(columns={'Trans. Date': 'Date'},inplace=True)
discoverTransactions.head()
```

```python
bofaCashRewardsTransactions['Account']='BOfACashRewards'
bofaTravelRewardsTransactions['Account']='BOfATravelRewards'
bofaAccountTransactions['Account']='BOfAAccounts'
discoverTransactions['Account']='Discover'
```

```python
transactions = pd.concat([
    bofaCashRewardsTransactions[['Date','Account','Description','Amount','Category']]
    ,bofaTravelRewardsTransactions[['Date','Account','Description','Amount','Category']]
    ,bofaAccountTransactions[['Date','Account','Description','Amount','Category']]
    ,discoverTransactions[['Date','Account','Description','Amount','Category']]
])
transactions.head()
```

```python
transactions.info()
```

```python
transactions.Date = pd.to_datetime(transactions.Date)
transactions.info()
```

```python
transactions['Month'] = transactions['Date'].dt.to_period('M')
transactions.Category.fillna('Unknown', inplace=True)
transactions
```

```python
transactions.info()
```

```python
monthlySummary = transactions.groupby(['Month','Category']).agg({'Amount':{'Total':'sum'}})
monthlySummary.columns = monthlySummary.columns.droplevel()
monthlySummary.reset_index(inplace=True)
monthlySummary.sort_values(['Month','Total'],ascending=False)
```
